```python
N = torch.zeros((28, 28), dtype = torch.int32)
```

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i for i, s in enumerate(chars)} #used in giving each letter a index
stoi['<S>'] = 26
stoi['<E>'] = 27
```

```python
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1] #the stoi of ch1
        ix2 = stoi[ch2] #the stoi of ch2
        N[ix1, ix2] += 1 # adds 1 to the ix1th row and ix2th column
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(N)
```
